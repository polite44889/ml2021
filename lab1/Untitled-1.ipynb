# 3. Matplotlib (створити набір даних лінійної функції з урахуванням помилок вимірювань, побудувати графік функції та згенерованих даних, обчислити похибку з використанням метрик MAE, MSE). Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

```python
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
x = np.arange(0, 6, 1)
y1 = x * 2

adder = np.random.random(len(x))-0.2
y2 = y1 + adder

plt.figure()
plt.grid()
plt.xticks(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x, y1, "r--", label="teoretic")
plt.plot(x, y2, "b", label="result")

plt.legend()
plt.show()
```

```python
mae = metrics.mean_absolute_error(y2, y1)
mse = metrics.mean_squared_error(y2, y1)
print(mae)
print(mse)
```

```python
dfmat=pd.DataFrame({
    'X':x,
    'Y':y1,
    'Y_hat':y2,
    'MAE':mae,
    'MSE':mse

})
dfmat.to_csv('output.csv')
dfmat
```
